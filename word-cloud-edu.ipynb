```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt #to display our wordcloud
from matplotlib.pyplot import figure
from PIL import Image #to load our image
import numpy as np #to get the color of our image
import nltk
```

```python
df = pd.read_csv('dataset/vacina.csv', encoding = "cp1252")
```

```python
df.head()
```

```python
df.shape
```

```python
word_list = list(df['Word'])
#word_list
```

```python
list_qty = list(df['Word Count'])
#list_qty
```

```python
general_list = []

for i in range(df.shape[0]):
  total_words = list_qty[i]*(',' + word_list[i])
  total_words = total_words.split(sep = ',')
  total_words.remove('')
  general_list.append(total_words)
```

```python
final_list = [item for sublist in general_list for item in sublist]
#final_list
```

```python
with open('dataset/my_table.txt', 'w') as file:
    file.write(', '.join(final_list))
```

```python
text = open('dataset/my_table.txt', 'r').read()
```

```python
#custom_mask = np.array(Image.open('img/image.png'))
```

```python
stopwords_pt = nltk.corpus.stopwords.words('portuguese')
```

```python
# parameters for plotting
wc = WordCloud(
    stopwords=stopwords_pt,
    collocations = False,
    max_words=1000,
    colormap='gray',
    font_path='font/NotoSerif-Bold.ttf',
    #min_font_size=4,
    max_font_size=600,
    #regexp='',
    #mask=custom_mask,
    #contour_color='black',
    #contour_width=5,
    #prefer_horizontal=0.90,
    width= 1600,
    height=800,
    background_color="white"
)
```

```python
wc.generate(text) # Gera a plotagem
```

```python
plt.imshow(wc, interpolation='bilinear') # Parametro de exibição
plt.axis("off") # Parametro de exibição
plt.show() # Exibe a nuvem de palavras

wc.to_file('export/wc-vacina.png') # Salva uma cópia da nuvem de palavras 
```
